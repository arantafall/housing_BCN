```python
import numpy as np
import seaborn as sns
import pandas as pd
df = pd.read_csv('../data/Barcelona_Fotocasa_HousingPrices.csv')
```

```python
# creem una matriu de correlacio amb totes les variables numriques (preu, habitacions, bany, metres cuadrats i preu x metre cuadrat) per veure quines variables tenen relaci directe
matriu_correlacio = df[['price', 'rooms', 'bathroom', 'square_meters', 'square_meters_price']].corr()
sns.heatmap(matriu_correlacio, annot=True, cmap='coolwarm')
```

```python
# Detectar relacions no lineals o patrons interessants entre variables.
sns.pairplot(df, vars=['price', 'rooms', 'bathroom', 'square_meters', 'square_meters_price'], diag_kind='kde')
```

```python
preu_mitja_barri = df.groupby('neighborhood')['price'].mean().sort_values(ascending=False)
print(preu_mitja_barri)
```

```python
sns.boxplot(x='real_state', y='price', data=df)
```
