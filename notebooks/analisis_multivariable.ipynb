```python
import numpy as np
import seaborn as sns
import pandas as pd
df = pd.read_csv('../data/Barcelona_Fotocasa_HousingPrices.csv')
```

```python
# creem una matriu de correlacio amb totes les variables numériques (preu, habitacions, bany, metres cuadrats i preu x metre cuadrat) per veure quines variables tenen relació directe
matriu_correlacio = df[['price', 'rooms', 'bathroom', 'square_meters', 'square_meters_price']].corr()
sns.heatmap(matriu_correlacio, annot=True, cmap='coolwarm')
```

```python
# Detectar relacions no lineals o patrons interessants entre variables (numériques).
sns.pairplot(df, vars=['price', 'rooms', 'bathroom', 'square_meters', 'square_meters_price'], diag_kind='kde')
```

```python
# Preu mitjà de cada barri
df.groupby('neighborhood')['price'].mean().sort_values(ascending=False)
```

```python
# Preu segons el tipus de vivenda
sns.boxplot(x='real_state', y='price', data=df)
```
