```python
data = data.dropna(subset=['property_type'])
data.reset_index(drop=True, inplace=True)
```

```python
print(data[['square_meters_price', 'neighborhood']].describe())
```

```python
print(data['neighborhood'].value_counts())
```

```python
data['square_meters_price'].describe();
data['neighborhood'].describe()
```

```python
data['square_meters_price'] > 1000
```

```python
#Precio por metro cuadrado depeniendo del barrio
grouped_stats = data.groupby('neighborhood')['square_meters_price'].mean()
(grouped_stats)
```

```python
#Precio por m2 dependiendo del  barrio
price_stats = data.groupby('neighborhood')['square_meters_price'].agg(['mean', 'min', 'max', 'median']).round(2)
(price_stats)
```

```python
price_per_room = data.groupby('rooms')['square_meters_price'].mean().round(2)
(price_per_room)
```

```python
price_by_features = data.groupby('lift')['square_meters_price'].mean()
(price_by_features)
```

```python
#Anlasi aument de precios dependiendo de numero de ba√±os
variacion = data.groupby(['neighborhood', 'bathroom'])['price'].mean().reset_index(). round(2)
(variacion)
```

```python
#Anlasi aument de precios dependiendo de numero de habitaciones
variacion = data.groupby(['neighborhood', 'rooms'])['price'].mean().reset_index(). round(2)
(variacion)

```

```python
#Anlasi aument de precios dependiendo de si tiene  acensor o no

variacion = data.groupby(['neighborhood', 'lift'])['price'].mean().reset_index(). round(2)
(variacion)
```

```python
#Anlasi aument de precios dependiendo de si tiene  terraza o no
variacion = data.groupby(['neighborhood', 'terrace'])['price'].mean().reset_index(). round(2)
(variacion)
```

```python
#Precio promedio y distribuci√≥n de precios por barrio
price_stats = data.groupby('neighborhood')['price'].agg(['mean', 'min', 'max', 'median']).round(2)
(price_stats)
```
